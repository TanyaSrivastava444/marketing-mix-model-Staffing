```python
import pandas as pd

# Replace 'your_file.csv' with the actual name of your uploaded CSV file
df_csv = pd.read_csv('dataset_synthetic.csv')

# Display the first 5 rows of the DataFrame
display(df_csv.head())
```

```python
#convert week to datetime
df_csv['week'] = pd.to_datetime(df_csv['week'])
```

```python
#sorting rows by date for adstock recursion function that calculates carryover effect
df_csv = df_csv.sort_values('week')
```

```python
#setting index - Makes week the row index. this is needed for time series modeling
df_csv = df_csv.set_index('week')

```

```python
#missing values using forward fill assuming spend remains constant until new value appears
df_csv = df_csv.fillna(method="ffill")
```

```python
# feautures list - apply transformations programmatically
channels = [
"job_board_spend",
"linkedin_spend",
"events_spend",
"youtube_ads_spend",
"google_ads_spend",
"client_outreach_spend"
]
```

```python
macro_vars = [
"unemployment_rate",
"job_openings",
"payrolls"
]
```

```python
## Scaling - transforms variables between 0-1  no scaling will throw gradient descent off
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

df_csv[channels] = scaler.fit_transform(df_csv[channels])
df_csv[macro_vars] = scaler.fit_transform(df_csv[macro_vars])
```
